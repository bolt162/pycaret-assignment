```python
# ============================================================
# DATA LOADING - NSL-KDD Network Intrusion Detection (Online)
# ============================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

print("ğŸ“¥ Downloading NSL-KDD dataset from GitHub...")

# Direct URL to NSL-KDD training set
url = "https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTrain%2B.txt"

# NSL-KDD column names (41 features + label)
column_names = [
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes',
    'dst_bytes', 'land', 'wrong_fragment', 'urgent', 'hot',
    'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell',
    'su_attempted', 'num_root', 'num_file_creations', 'num_shells',
    'num_access_files', 'num_outbound_cmds', 'is_host_login',
    'is_guest_login', 'count', 'srv_count', 'serror_rate',
    'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate',
    'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count',
    'dst_host_srv_count', 'dst_host_same_srv_rate',
    'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
    'dst_host_srv_diff_host_rate', 'dst_host_serror_rate',
    'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
    'dst_host_srv_rerror_rate', 'class', 'difficulty_level'
]

# Load dataset directly from URL
data_raw = pd.read_csv(url, names=column_names, header=None)
print(f"âœ… Dataset loaded: {data_raw.shape}")

# Convert attack labels to binary (0=normal, 1=attack)
y_true = (data_raw['class'] != 'normal').astype(int)

# Drop target and difficulty columns
data = data_raw.drop(columns=['class', 'difficulty_level'])

# Encode categorical features (protocol_type, service, flag)
print("ğŸ”„ Encoding categorical features...")
categorical_cols = ['protocol_type', 'service', 'flag']
for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))

# Optional: Sample for faster experimentation (recommended for first run)
SAMPLE_SIZE = 60000  # Set to None to use full dataset
if SAMPLE_SIZE and SAMPLE_SIZE < len(data):
    print(f"ğŸ“Š Sampling {SAMPLE_SIZE} records for faster processing...")
    sample_idx = data.sample(n=SAMPLE_SIZE, random_state=123).index
    data = data.loc[sample_idx].reset_index(drop=True)
    y_true = y_true.loc[sample_idx].reset_index(drop=True)

print("\n" + "="*60)
print("ğŸ“Š DATASET SUMMARY")
print("="*60)
print(f"Shape: {data.shape}")
print(f"\nLabel distribution:")
print(y_true.value_counts().to_frame('count'))
print(f"\nClass balance: {y_true.mean()*100:.2f}% attacks")
print(f"\nFeature types:")
print(data.dtypes.value_counts())
print("\n" + "="*60)

data.head()
```

```python
# PyCaret Anomaly setup (PyCaret 3.x)
from pycaret.anomaly import (
    setup, create_model, assign_model, plot_model,
    predict_model, save_model, load_model, models
)

s = setup(
    data=data,          # features only (no 'Class')
    session_id=123,     # reproducibility
    normalize=True,     # scale V-features & Amount
    transformation=False,
    verbose=True        # keep logs; set to False for quieter output
)

s  # display experiment summary
```

```python
# List available anomaly models (optional)
models()

# Train a few solid baselines
iforest = create_model('iforest')   # Isolation Forest
knn     = create_model('knn')       # KNN outlier detector
lof     = create_model('lof')       # Local Outlier Factor

# (Optional extras you can try later)
# hbos = create_model('hbos')
# copod = create_model('copod')
# pca   = create_model('pca')
```

```python
iforest_results = assign_model(iforest)
knn_results     = assign_model(knn)
lof_results     = assign_model(lof)

# quick peek
iforest_results.head()
```

```python
pred_iforest = predict_model(iforest, data=data)
pred_knn     = predict_model(knn, data=data)
pred_lof     = predict_model(lof, data=data)

pred_iforest.head()
```

```python
# Make sure y_true matches the same rows as `data`.
# If you downsampled earlier, ensure y_true was subset to the same index.
assert len(y_true) == len(data), "y_true length must match the current `data` length."

import numpy as np, pandas as pd
from sklearn.metrics import roc_auc_score, average_precision_score, precision_recall_fscore_support, confusion_matrix

def evaluate_predictions(y_true_arr, pred_df, name):
    y_pred  = pred_df['Anomaly'].astype(int).values
    scores  = pred_df['Anomaly_Score'].values
    auc     = roc_auc_score(y_true_arr, scores)
    ap      = average_precision_score(y_true_arr, scores)
    p, r, f1, _ = precision_recall_fscore_support(y_true_arr, y_pred, average='binary', zero_division=0)
    cm      = confusion_matrix(y_true_arr, y_pred)

    print(f"=== {name} ===")
    print(f"AUC-ROC: {auc:.4f} | PR-AUC: {ap:.4f}")
    print(f"Precision: {p:.4f} | Recall: {r:.4f} | F1: {f1:.4f}")
    print("Confusion Matrix [ [TN FP]\n                     [FN TP] ]")
    print(cm, "\n")
    return {'model': name, 'auc': auc, 'ap': ap, 'precision': p, 'recall': r, 'f1': f1}

eval_rows = []
eval_rows.append(evaluate_predictions(y_true.values, pred_iforest, "Isolation Forest"))
eval_rows.append(evaluate_predictions(y_true.values, pred_knn, "KNN"))
eval_rows.append(evaluate_predictions(y_true.values, pred_lof, "LOF"))

eval_df = pd.DataFrame(eval_rows).sort_values('ap', ascending=False).reset_index(drop=True)
eval_df
```

```python
import numpy as np
from sklearn.metrics import precision_recall_fscore_support

def tune_threshold(scores, y_true_arr, target="f1", target_recall=None):
    # Try percentiles to choose a threshold on the score (higher = more anomalous)
    percentiles = np.linspace(90, 99.9, 50)  # adjust wider/narrower as you like
    best = {"thr": None, "p": 0, "r": 0, "f1": 0}
    for q in percentiles:
        thr = np.percentile(scores, q)
        y_pred = (scores >= thr).astype(int)
        p, r, f1, _ = precision_recall_fscore_support(y_true_arr, y_pred, average='binary', zero_division=0)
        if target_recall is not None:
            # choose highest precision meeting min recall
            if r >= target_recall and p > best["p"]:
                best = {"thr": thr, "p": p, "r": r, "f1": f1}
        else:
            # maximize the target metric
            val = {"precision": p, "recall": r, "f1": f1}[target]
            if val > best[target]:
                best = {"thr": thr, "p": p, "r": r, "f1": f1}
    return best

scores = pred_iforest["Anomaly_Score"].values
best = tune_threshold(scores, y_true.values, target="f1")  # or target_recall=0.80
best
```
